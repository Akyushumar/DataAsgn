```python
import pandas as pd

# List of Excel files to combine
files = ["BedD.xlsx", "ED.xlsx", "IAQ.xlsx", "OAQ.xlsx"]

# Create a new Excel writer object
with pd.ExcelWriter("Mastersheet.xlsx", engine="openpyxl") as writer:
    for file in files:
        df = pd.read_excel(file)  # Read each file
        sheet_name = file.split(".")[0]  # Use file name as sheet name
        df.to_excel(writer, sheet_name=sheet_name, index=False)  # Save in new sheet

print("Files merged successfully!")
```
